```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("train.csv")
```

```python
# 1. Basic Information
print("Dataset Info:")
print(df.info())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
# 2. Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# 3. Survival Count
print("\nSurvival Count:")
print(df['Survived'].value_counts())
```

```python
# 4. Survival by Gender
print("\nSurvival by Gender:")
print(df.groupby(['Sex', 'Survived']).size())
```

```python
# 4. Survival by Gender
print("\nSurvival by Gender:")
print(df.groupby(['Sex', 'Survived']).size().unstack())
```

```python
# 5. Survival by Passenger Class
print("\nSurvival by Class:")
print(df.groupby(['Pclass', 'Survived']).size().unstack())
```

```python
# 6. Distribution of Age
#This is a histogram of Age distribution from the Titanic dataset.
#X-axis → Age
#Y-axis → Count (Number of passengers)
#Each bar represents how many passengers fall within an age range (bin).

plt.figure()
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

From the above Histogram we can interpret:
Majority of passengers were young adults

* Most passengers are between 20–40 years.

* The peak appears around 25–30 years.
* Very few passengers above 60 years.

* Count gradually decreases as age increases.
* Small number of passengers between 0–10 years.



```python
# 7. Fare Distribution
plt.figure()
df['Fare'].hist(bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# 8. Correlation Matrix
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

# Correlation Matrix shows the relationship between numerical variables.

Values range from -1 to +1

+1 → Strong positive correlation

-1 → Strong negative correlation

 0 → No relationship

 "The correlation matrix shows that passenger class has a moderate negative correlation with survival, meaning first-class passengers had better survival chances. Fare has a weak positive correlation with survival, while age has very little correlation."
