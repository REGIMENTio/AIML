```python
#Load Dataset
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

# Step 2: Display number of rows and columns

```python
print("Shape of Dataset (Rows, Columns):", df.shape)
print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

# Step 3: Display first five rows


```python
print("\nFirst Five Rows:")
print(df.head())
```


# Step 4: Display dataset size (total elements)

```python
print("\nSize of Dataset:", df.size)
```

# Step 5: Check missing values

```python
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

# Step 6: Select numerical columns

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
print("Number of Rows:", numeric_df.shape[0])
print("Number of Columns:", numeric_df.shape[1])
```

# Step 7: Display statistical values

```python
print("\nSum of Numerical Columns:")
print(numeric_df.sum())
```

```python
print("\nAverage of Numerical Columns:")
print(numeric_df.mean())

print("\nMinimum Values:")
print(numeric_df.min())

print("\nMaximum Values:")
print(numeric_df.max())
```

# Step 8: Export dataset

```python
df.to_csv("titanic_output.csv", index=False)

print("\nDataset exported successfully as 'titanic_output.csv'")
```
