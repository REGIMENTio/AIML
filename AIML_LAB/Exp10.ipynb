```python
import matplotlib.pyplot as plt
import numpy as np
# ── FIGURE ANATOMY ─────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
# fig = the whole window
# ax  = the individual plot area
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 7])
ax.plot(x, y,
        color='purple',       # line colour
        linewidth=2,          # line thickness
        marker='o',           # dot at each point
        markersize=8,
        label='My Data')
ax.set_title('My First Plot', fontsize=16, fontweight='bold')
ax.set_xlabel('X Axis Label', fontsize=12)
ax.set_ylabel('Y Axis Label', fontsize=12)
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()  # prevents overlap
plt.savefig('myplot.png', dpi=150)  # save to file
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()

```

```python
# load titanic dataset
df = pd.read_csv('train.csv')

# handle missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# convert Survived to category
df['Survived'] = df['Survived'].map({0:'Not Survived', 1:'Survived'})

# ── SET STYLE & PALETTE ────────────────────────
sns.set_theme(style='whitegrid')
sns.set_palette('viridis')

# ── 1. HISTOGRAM + KDE ────────────────────────
sns.histplot(data=df, x='Age',
             kde=True, hue='Survived')
plt.title('Age Distribution by Survival')
plt.show()

# ── 2. BOX PLOT (grouped) ─────────────────────
sns.boxplot(data=df, x='Survived', y='Fare',
            palette='Set2')
plt.title('Fare by Survival')
plt.show()

# ── 3. VIOLIN PLOT ────────────────────────────
sns.violinplot(data=df, x='Survived', y='Age')
plt.title('Age Distribution by Survival (Violin)')
plt.show()

# ── 4. HEATMAP ────────────────────────────────
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True, fmt='.2f',
            cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# ── 5. PAIRPLOT ───────────────────────────────
sns.pairplot(df[['Age','Fare','Pclass','SibSp','Parch','Survived']],
             hue='Survived',
             diag_kind='kde')
plt.suptitle('Titanic Pairplot', y=1.02)
plt.show()



```
